```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv("datasets/Brain-BodySize/brain-bodysize.csv")
df1.head(200)
```

```python
X= df1[['Brain']]
Y= df1[['Body']]

X.head()
```

```python
#using scikit-learn to train your model with LinearRegression.
model = sklearn.linear_model.LinearRegression()
model.fit(X,Y)
predict_Y = model.predict(X)
```

```python
#use matplotlib (or any visualization libraries) to make the plot describing this data.
#add your regression line to your plot, if you haven't already done so
df1.plot(kind='scatter', x="Brain", y="Body")
plt.plot(X,predict_Y)

plt.show()
```

```python
#create three different new unseen data points (Xs) and make predictions from your model (Ys)
new_x1 = model.predict([[350]])
new_x2 = model.predict([[550]])
new_x3 = model.predict([[650]])

print("predictions = ", new_x1, new_x2, new_x3)
```

```python
#(bonus points) Can you use other techniques to improve the accuracy of this model?

# I will calculate the root mean squared error using SKLearn analytics
RMSE = sklearn.metrics.mean_squared_error(Y, predict_Y)
print(RMSE)
#This is large. Let's reduce the error of our model
```

```python
#Cull outlier from data
df1_cull = df1.drop(30)
df1_cull = df1_cull.reset_index()
del df1_cull['index']
df1_cull.head(31)
```

```python
#retrain model with new data
cull_X = df1_cull[['Brain']]
cull_Y = df1_cull[['Body']]
model2 = sklearn.linear_model.LinearRegression()
model2.fit(cull_X,cull_Y)
predict_cull_Y = model2.predict(cull_X)
```

```python
#Visualize new data
df1_cull.plot(kind='scatter', x="Brain", y="Body")
plt.plot(cull_X,predict_cull_Y)
plt.show()
```

```python
#Now that the data is better visualized and fits the graph frame better, we recalculate the
#Mean Squared error of our regression

RSME_2 = sklearn.metrics.mean_squared_error(cull_Y, predict_cull_Y)
print(RSME_2)
```
